```python
## all util functions

# get bib_key from input string    
def get_key_from_bib(bib_entry):
    bib_data = pbx.parse_string(bib_entry, "bibtex")
    return list(bib_data.entries.keys())[0]

# check if key is duplicate
def is_key_dulicates(bib_key):
    keys_list = get_list_stored_keys()
    if bib_key in keys_list:
        return 1
    else:
        return 0
    
# read from the list of stored bib_keys
def get_list_stored_keys():
    keys_list = []
    with open('stored_bib_keys.txt', 'r') as filehandle:
        for line in filehandle:
            if line.strip(): # skip empty lines
                keys_list.append(line[:-1])
    return keys_list

# append key to file
def append_key_to_file(bib_key):
    with open('stored_bib_keys.txt', 'a') as filehandle:
        filehandle.write('%s\n' % bib_key)
        
# append csv file for keeping track of category of bib_key
def update_bib_csv(bib_key,categories):  
    line=[bib_key]
    line.extend(categories)
    line = ','.join(map(str, line))
    with open('bib_key_caetgory.csv', 'a') as filehandle:
        filehandle.write('%s\n' % line)    
```
